# Importando a base de dados

```python
import pandas as pd
dados = pd.read_csv('arquivo/tips.csv')
dados
```

## Tradução da Base de Dados

```python
dados.columns
```

```python
renomear = {
    'total_bill' :' valor_da_conta', 
    'tip' : 'gorjeta', 
    'dessert' : 'sobremesa',
    'day' : 'dia_da_semana', 
    'time' : 'hora_do_dia', 
    'size' : 'total_de_pessoas'
}
```

```python
gorjeta = dados.rename(columns = renomear)
gorjeta.head(5)
```

```python
gorjeta.sobremesa.unique()
```

```python
sim_nao = {
    'No' : 'Não', 
    'Yes' : 'Sim'
}
```

```python
gorjeta.sobremesa.map(sim_nao)
```

```python
gorjeta.sobremesa = gorjeta.sobremesa.map(sim_nao)
gorjeta.sample(4)
```

```python
gorjeta.dia_da_semana.unique()
```

```python
dias = {
    'Sun' : 'Domingo', 
    'Sat' : 'Sábado', 
    'Thur' : 'Quita-Feira',
    'Fri' : 'Sexta-Feira'
}
```

```python
gorjeta.dia_da_semana = gorjeta.dia_da_semana.map(dias)
gorjeta.sample(4)
```

```python
gorjeta.hora_do_dia.unique()
```

```python
hora = {
    'Dinner' : 'Jantar', 
    'Lunch' : 'Almoço'
}
```

```python
gorjeta.hora_do_dia = gorjeta.hora_do_dia.map(hora)
```

```python
gorjeta.sample(10)
```
